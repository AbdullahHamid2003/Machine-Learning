## To Predict the prices of the year 2020 based on the given dataset 

```python
import pandas as pd 
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
```

```python
df.head()
```

```python
df.rename(columns={'per capita income (US$)': 'Income (US$)'}, inplace=True)
```

```python
df.head()
```

```python
plt.xlabel('year')
plt.ylabel('Income (US$)')
plt.scatter(df['year'], df['Income (US$)'],color='red',marker='+')
plt.title('Income Over Years')  
plt.show()
```

```python
# creating linear regression object
reg = linear_model.LinearRegression()
```

# Training the model

```python
X = df[['year']]  # Features should be 2D
y = df['Income (US$)']  # Target should be 1D
reg.fit(X, y)
```

```python
reg.predict([[2020]])
```
